```python
import random as r
import string as s

def random_string(length):
    return ''.join(r.choice(s.ascii_letters) for _ in range(length))

# Generate random strings and write them to a file
with open('random_strings.txt', 'w') as file:
    for _ in range(1000):
        random = random_string(5)  
        file.write(random + '\n')

print("File 'random_strings.txt' created")
```

```python
import random as r
import string as s

def random_string(length):
    return ''.join(r.choice(s.ascii_letters) for _ in range(length))

# Open a file for writing
with open('random_strings1.txt', 'w') as file:
    file_size = 0
    max_size = 5*1024*1024
    while file_size <= max_size: #5MB
        # Generate a random string of random length 
        random_length = r.randint(1,5)
        random = random_string(random_length)
 # Write to the file
        file.write(random + '\n')

        # Update the file size
        file_size = file.tell()

print("File 'random_strings1.txt' created")
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
dataset = pd.DataFrame(np.random.randint(1, 201, size=(100, 30)))
print(dataset)
dataset = dataset.mask((dataset >= 10) & (dataset <= 60), np.nan)
rows_with_missing_values = dataset.isnull().sum(axis=1)
count_rows_with_missing_values = rows_with_missing_values[rows_with_missing_values > 0].count()
print(count_rows_with_missing_values)
column_averages = dataset.mean()
dataset = dataset.fillna(column_averages)
count_rows_with_missing_values = rows_with_missing_values[rows_with_missing_values > 0].count()
print(count_rows_with_missing_values)
correlation_matrix = dataset.corr()
plt.figure()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Pearson Correlation Heatmap')
plt.show()
#columns_to_select = correlation_matrix.columns[correlation_matrix.abs() <= 0.7].tolist()

#print("Columns with correlation <= 0.7:")
#print(columns_to_select)

```
